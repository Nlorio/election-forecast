```python
house_names = pd.read_csv(DATA_DIR/'cleaned'/'candidate_parties_house.csv')
print(house_names.shape)
house_names.head()
```

```python
house_map = house_names[['party']]
house_map.index = (
    house_names['race_type'] + 
    '_' + 
    house_names['name'].str.split().map(lambda x: x[-1]) + 
    '_' + 
    house_names['state']
)
house_map = house_map.to_dict()['party']
```

```python
def expand_polls_house(data):
    columns = ['candidate_name', 'candidate_party', 
               'race_name', 'race_type', 'state',
               'candidate_poll', 'opponent_poll',
               'leading_poll', 'date', 'win'
              ]
    new = pd.DataFrame({'candidate_name' : [candidate for sublist in data['top_candidates'] for candidate in sublist],
                        'candidate_party' : '', # Filled in below
                        'race_name' : data.loc[data.index.repeat(2), 'race_name'],
                        'race_type' : data.loc[data.index.repeat(2), 'race_type'],
                        'state' : data.loc[data.index.repeat(2), 'state'],
                        'district' : data.loc[data.index.repeat(2), 'district'],
                        'candidate_poll' : -1, # Filled in below
                        'opponent_poll' : -1, # Filled in below
                        'leading_poll' : False, # Filled in below
                        'poll_results' : data.loc[data.index.repeat(2), 'poll_results'],
                        'date' : data.loc[data.index.repeat(2), 'date'],
                        'election_date' : data.loc[data.index.repeat(2), 'election_date'],
                       }, index=data.index.repeat(2))
    new = new.reset_index(drop=True)
    
    # map candidate party
    new['candidate_id'] = (new['race_type'] + '_' + new['candidate_name']
                              .str.replace('-', '')
                              .str.replace(r'\.', '')
                              .str.replace(',', '')
                              .str.replace(r'\'', '')
                              + '_' + new['state']
                          + '_' + new['date'].dt.year.astype(str)[:4])
    new['candidate_party'] = new['candidate_id'].map(house_map)
    
    def find_opp(row):
        keys = row['poll_results'].keys()
        keys = list(keys)
        keys.remove(row['candidate_name'])
        return row['poll_results'][keys[0]]
    
    new['candidate_poll'] = new.apply(lambda x: x['poll_results'][x['candidate_name']], axis=1)
    new['opponent_poll'] = new.apply(lambda x: find_opp(x), axis=1)
    new['spread'] = new['candidate_poll'] - new['opponent_poll']
    new['leading_poll'] = new['spread'] > 0
    
    return new
```

```python
house_polls_tr = expand_polls_house(house_polls)
```

# Feature Engineering

```python
class ExtractHouseFeatures(BaseEstimator, TransformerMixin):
    def __init__(self, polls):
        self.polls = polls
        pass
    
    def fit(self, X, y=None):
        return self
    
    def transform(self, X, y=None):
        X['average_spread'] = -1
        X['weighted_spread'] = -1
        
        for _, row in X.iterrows():
            name = row['candidate_id']
            subset = self.polls[self.polls['candidate_id'] == name]
            print(subset)
            break
        return X
```
