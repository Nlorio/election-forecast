# Importing Packages

```python
import numpy as np
import pandas as pd
import re
import requests
from collections import Counter
import bs4 as bs
```

# 112th Congress (from January 3, 2011, until January 3, 2013)

```python
source = requests.get("https://en.wikipedia.org/wiki/112th_United_States_Congress#Members")
soup = bs.BeautifulSoup(source.content, features='html.parser') 
begins = [str(link.get('href')) for link in soup.find_all('a')]
ends = []
for string in begins:
    if re.match(r"/wiki", string):
        ends.append(re.sub(r"/wiki/","",string))
sens112 = ends[237:390]
reps112 = ends[400:1331]
senate112 = []
for sen in sens112:
    if re.match(r"List_of.*", sen):
        state = re.sub(r"^List_of_United_States_Senators_from_","STATE: ",sen)
    else:
        senate112.append(state)
        name = re.sub(r"_"," ", sen)
        senate112.append("SENATOR: " + re.sub(r"\(.*\)","",name))
house112 = []
for dis in reps112:
    if re.match(r"List_of.*", dis):
        state = re.sub(r"^List_of_United_States_Representatives_from_","STATE: ",dis)
    elif re.match(r".*_congressional_district$", dis):
        house112.append(state)
        temp = re.sub(r"_congressional_district$","",dis)
        house112.append("DISTRICT: " + re.sub(r"^.*[^_]+_","", temp))
    else:
        name = re.sub(r"_"," ",dis)
        house112.append("REP: " + re.sub(r"\(.*\)","",name))
```

```python
senate112
```

```python
house112
```
