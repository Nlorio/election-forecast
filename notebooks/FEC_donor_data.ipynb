## Feature Extraction 

Objective of aggregation:

    Unique row for:
        Each candidate 
        Sum of donations per feature as long as the value being summed is not equal to the other rows values. 
        List of committees that funded.
        
    Features: 
        Number of committees funding
        Incumbent vs. Challenger
        State Donations
        Donation Levels
 

```python
donor_df_grouped = donor_df.groupby(['name', 'cand_id', 'cycle', 'state', 
                                     'incumbent', 'office_full', 'party', 'committee_id']).max()
```

```python
# Remove Duplicates
# 
donor_df_group = donor_df_grouped.reset_index()
```

```python


def clean_dataset(donor_df_group):
    features = ['Armed Forces Americas', 'Armed Forces Europe',
       'Alaska', 'Alabama', 'Armed Forces Pacific', 'Arkansas',
       'American Samoa', 'Arizona', 'California', 'Colorado', 'Connecticut',
       'District Of Columbia', 'Delaware', 'Florida', 'Georgia', 'Guam',
       'Hawaii', 'Iowa', 'Idaho', 'Illinois', 'Indiana', 'Kansas', 'Kentucky',
       'Louisiana', 'Massachusetts', 'Maryland', 'Maine', 'Michigan',
       'Minnesota', 'Missouri', 'Northern Mariana Islands', 'Mississippi',
       'Montana', 'North Carolina', 'North Dakota', 'Nebraska',
       'New Hampshire', 'New Jersey', 'New Mexico', 'Nevada', 'New York',
       'Ohio', 'Oklahoma', 'Oregon', 'Other', 'Pennsylvania', 'Puerto Rico',
       'Rhode Island', 'South Carolina', 'South Dakota', 'Tennessee', 'Texas',
       'Utah', 'Virginia', 'U.S. Virgin Islands', 'Vermont', 'Washington',
       'Wisconsin', 'West Virginia', 'Wyoming', 'Foreign Countries',
       'Donation Level 1', 'Donation Level 2', 'Donation Level 3',
       'Donation Level 4', 'Donation Level 5']
     
    df_temp = donor_df_group.copy()
    cols = df_temp.columns

    df = pd.DataFrame(np.full((len(df_temp.name), len(cols)), np.nan), columns = cols)
    j = 0
    for name in np.unique(df_temp.name.values):
        print(name)
        df_name = df_temp[df_temp.name == name]
        funding_committees = [x for x in df_name.committee_name if x != 0]
        length = len(funding_committees)
        if length == 0:
            funding_committees = 0
#         race_candidates = [x for x in df_name.race_candidates if x != 0]
#         election_winner = [x for x in df_name.election_winner if x != 0]
#         incumbent  = [x for x in df_name.incumbent if x != 0]
#         office_full = [x for x in df_name.office_full if x != 0]
#         party = [x for x in df_name.party if x != 0]
        feature_donations = []
        for feature in features:
            unique_features = np.unique(df_name[feature].values)
            feature_donations = np.append(feature_donations, np.sum(unique_features))

        #unique row for each candidate name
        name_row = df_name[df_name.office_full != 0]
        if name_row.empty:
            continue
        df.iloc[j, :] = name_row.iloc[0, :]
        df['committee_name'] = df['committee_name'].astype(list)
        
        
        df.iloc[j, 7] = length
        df.iat[j, 8] = funding_committees
        
        i = 11
        for feature_donation in feature_donations:
            df.iloc[j, i] = feature_donation
            i += 1
        j += 1 
        print('{}/614'.format(j))
#         if j == 10:
#             break
    return df.dropna()
        

```

```python
df_donor_clean = clean_dataset(donor_df_group)
```
