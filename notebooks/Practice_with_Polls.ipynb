# Importing Packages

```python
import numpy as np
import pandas as pd
import re
```

# Scraping 

```python
today = pd.read_html("https://www.realclearpolitics.com/epolls/latest_polls")
for i in range(len(today)):
    if i%2==1: #looks at every other object in the list
        today[i].columns = list(today[i].iloc[0,:])
        today[i] = today[i].drop(0)
        today[i]["Date"] = today[i-1][0][0]
today = pd.concat([today[x] for x in range(len(today)) if x%2==1]) #joins all the separate dataframes
today = today.reset_index().drop('index', axis = 1) #sets new index and drops the old one
#today.to_csv("rcptoday.csv")
#run above line uncommented to save as .csv file
```

```python
senate = pd.DataFrame(columns=['Race/Topic (Click to Sort)','Poll','Results','Spread', 'Date'])
for element in range(len(today)):
    race = today.iloc[element,:]["Race/Topic (Click to Sort)"]
    z = re.match(r".*[Ss]enate.*", race)
    if z:
        senate = senate.append(today.iloc[element,:], ignore_index = True)
senate.head()
```
