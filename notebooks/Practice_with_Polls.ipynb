```python
topics = today["Race/Topic (Click to Sort)"].unique().tolist() #these are all the unique rows
qs = []
for race in topics:
    a = re.match(r".*[Dd]irection.*", race)
    b = re.match(r".*[Gg]eneric.*", race)
    c = re.match(r".*[Aa]pproval.*", race)
    d = re.match(r".*[Gg]overnor.*", race)
    e = re.match(r".*[Dd]istrict.*", race)
    f = re.match(r".*[Ss]enate.*", race)
    if a or b or c or d or e or f:
        None
    else:
        print(race)
```

```python
senate = pd.DataFrame(columns=['Race/Topic (Click to Sort)','Poll','Results','Spread', 'Date'])
for element in range(len(today)):
    race = today.iloc[element,:]["Race/Topic (Click to Sort)"]
    z = re.match(r".*[Ss]enate.*", race)
    if z:
        senate = senate.append(today.iloc[element,:], ignore_index = True)
senate.head()
```
