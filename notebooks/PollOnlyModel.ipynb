# Reading in data

```python
polls = pd.read_csv("../data/cleaned/Final_Polls_Nov_5.csv").drop("Unnamed: 0", axis = 1)
house_results = pd.read_csv("../data/cleaned/Clean_House_Winners.csv").drop("Unnamed: 0", axis = 1)
senate_results = pd.read_csv("../data/cleaned/Clean_Senate_Winners.csv").drop("Unnamed: 0", axis = 1)
governor_results = pd.read_csv("../data/cleaned/Clean_Governors_Final.csv").drop("Unnamed: 0", axis = 1)
```

# Data manipulation

```python
polls.columns = ["Race", "Poll", "Results", "Spread", "Date", "Year"]
```

```python
senate = pd.DataFrame(columns=['Race','Poll','Results','Spread', 'Date'])
for element in range(len(polls)):
    race = polls.iloc[element,:]["Race"]
    z = re.match(r".*[Ss]enate.*", race)
    if z:
        senate = senate.append(polls.iloc[element,:], ignore_index = True)
senate["Race"] = senate["Race"].str.lower()
senate["State"] = pd.Series([re.findall("^[a-z]+",x)[0] for x in senate["Race"]])
senate.head()
```

```python
house = pd.DataFrame(columns=['Race','Poll','Results','Spread', 'Date'])
for element in range(len(polls)):
    race = polls.iloc[element,:]["Race"]
    z = re.match(r".*[Dd]istrict.*", race)
    if z:
        house = house.append(polls.iloc[element,:], ignore_index = True)
house["Race"] = house["Race"].str.lower()
house.head()
```
